## Metricas de Evaluacion

Cohen Kappa o Kappa Cuadratica

${\displaystyle \kappa \equiv {\frac {p_{o}-p_{e}}{1-p_{e}}}=1-{\frac {1-p_{o}}{1-p_{e}}},\!}$


donde  es el acuerdo relativo observado entre los evaluadores (id茅ntico a la precisi贸n), y  es el acuerdo hipot茅tico de probabilidad de azar, utilizando los datos observados para calcular las probabilidades de que cada observador vea aleatoriamente cada categor铆a.

Por ejemplo:

true_labels = np.array([1, 0, 1, 1, 0, 1])
pred_labels = np.array([1, 0, 0, 0, 0, 1])

==(○+○)/=2+26=0.66

tp (true positive)
true negative

$p_{\text {yes }}=\frac{t p+f p}{\text { all }} \frac{t p+f n}{\text { all }}=\frac{2}{6} \frac{4}{6}=0.222$
$p_{\text {no }}=\frac{f n+\operatorname{tn} \frac{f p+\operatorname{tn}}{a l l}}{\text { all }}=\frac{4}{6} \frac{2}{6}=0.222$
$p_{e}=p_{\text {yes }}+p_{n o}=0.222+0.222=0.444$

Finalmente se calcula

$\kappa=\frac{p_{o}-p_{e}}{1-p_{e}}=\frac{0.666-0.444}{1-0.444}=0.4$


En pocas palabras, si dos puntajes no est谩n de acuerdo, la penalizaci贸n depender谩 de cu谩n lejos est茅n separados. Eso significa que nuestro puntaje ser谩 mayor si (a) el valor real es 4 pero el modelo predice un 3, y el puntaje ser谩 menor si (b) el modelo predice un 0.

## Preprocesamiento

```python
def get_pad_width(im, new_shape, is_rgb=True):
    pad_diff = new_shape - im.shape[0], new_shape - im.shape[1]
    t, b = math.floor(pad_diff[0]/2), math.ceil(pad_diff[0]/2)
    l, r = math.floor(pad_diff[1]/2), math.ceil(pad_diff[1]/2)
    if is_rgb:
        pad_width = ((t,b), (l,r), (0, 0))
    else:
        pad_width = ((t,b), (l,r))
    return pad_width

def preprocess_image(image_path, desired_size=224):
    im = Image.open(image_path)
    im = im.resize((desired_size, )*2, resample=Image.LANCZOS)
    
    return im
```

## Generador y Mezclador de imagenes


```python
class MixupGenerator():
    def __init__(self, X_train, y_train, batch_size=32, alpha=0.2, shuffle=True, datagen=None):
        self.X_train = X_train
        self.y_train = y_train
        self.batch_size = batch_size
        self.alpha = alpha
        self.shuffle = shuffle
        self.sample_num = len(X_train)
        self.datagen = datagen

    def __call__(self):
        while True:
            # barajea y crea los lotes 
            indexes = self.__get_exploration_order()
            itr_num = int(len(indexes) // (self.batch_size * 2))

            for i in range(itr_num):
                batch_ids = indexes[i * self.batch_size * 2:(i + 1) * self.batch_size * 2]
                X, y = self.__data_generation(batch_ids)

                yield X, y
    
    def __get_exploration_order(self):
        # indices de conjunto de entrenamiento
        indexes = np.arange(self.sample_num)
        
        #los revuelve
        if self.shuffle:
            np.random.shuffle(indexes)

        return indexes

    def __data_generation(self, batch_ids):
        _, h, w, c = self.X_train.shape
        l = np.random.beta(self.alpha, self.alpha, self.batch_size)
        X_l = l.reshape(self.batch_size, 1, 1, 1)
        y_l = l.reshape(self.batch_size, 1)

        X1 = self.X_train[batch_ids[:self.batch_size]]
        X2 = self.X_train[batch_ids[self.batch_size:]]
        X = X1 * X_l + X2 * (1 - X_l)

        if self.datagen:
            for i in range(self.batch_size):
                X[i] = self.datagen.random_transform(X[i])
                X[i] = self.datagen.standardize(X[i])

        if isinstance(self.y_train, list):
            y = []

            for y_train_ in self.y_train:
                y1 = y_train_[batch_ids[:self.batch_size]]
                y2 = y_train_[batch_ids[self.batch_size:]]
                y.append(y1 * y_l + y2 * (1 - y_l))
        else:
            y1 = self.y_train[batch_ids[:self.batch_size]]
            y2 = self.y_train[batch_ids[self.batch_size:]]
            y = y1 * y_l + y2 * (1 - y_l)

        return X, y
```

```python
def quadratic_kappa(actuals, preds, N=5):
    w = np.zeros((N,N))
    O = confusion_matrix(actuals, preds)
    print(O)
    for i in range(len(w)): 
        for j in range(len(w)):
            w[i][j] = float(((i-j)**2)/(N-1)**2)
    print(w)
    
    act_hist=np.zeros([N])
    for item in actuals: 
        act_hist[item]+=1
    
    pred_hist=np.zeros([N])
    for item in preds: 
        pred_hist[item]+=1

    print(act_hist)
    print(pred_hist)
                         
    E = np.outer(act_hist, pred_hist);
    print(E)
    E = E/E.sum();
    print('E')
    print(E)
    O = O/O.sum();
    print('O')
    print(O)

    num=0
    den=0
    for i in range(len(w)):
        for j in range(len(w)):
            num+=w[i][j]*O[i][j]
            den+=w[i][j]*E[i][j]
    return (1 - (num/den))
```

```python
def kappa_loss(y_pred, y_true, y_pow=2, eps=1e-10, N=5, bsize=34, name='kappa'):
  
  with tf.name_scope(name):
    y_true = tf.cast(y_true,dtype='float')
    repeat_op = tf.cast(tf.tile(tf.reshape(tf.range(0, N), [N, 1]), [1, N]), dtype='float')
    repeat_op_sq = tf.square((repeat_op - tf.transpose(repeat_op)))
    weights = repeat_op_sq / tf.cast((N - 1) ** 2, dtype='float')
    pred_ = y_pred ** y_pow

    try:
        pred_norm = pred_ / (eps + tf.reshape(tf.reduce_sum(pred_, 1), [-1, 1]))
    except Exception:
        pred_norm = pred_ / (eps + tf.reshape(tf.reduce_sum(pred_, 1), [bsize, 1]))

    hist_rater_a = tf.reduce_sum(pred_norm, 0)
    hist_rater_b = tf.reduce_sum(y_true, 0)

    conf_mat = tf.matmul(tf.transpose(pred_norm), y_true)

    nom = tf.reduce_sum(weights * conf_mat)
    denom = tf.reduce_sum(weights * tf.matmul(
        tf.reshape(hist_rater_a, [N, 1]), tf.reshape(hist_rater_b, [1, N])) /
                          tf.cast(bsize, dtype='float'))

    return  nom/(denom + eps)
```

```python
def quadratic_kappa_coefficient(y_true, y_pred):
  
    #y_pred = y_pred > 0.5
    #y_pred=tf.dtypes.cast(y_pred, tf.int32)
    #y_pred=tf.dtypes.cast(y_pred, "float32")
    y_pred = tf.round(y_pred)
    y_pred = K.sum(y_pred, axis=1)
    y_true= K.sum(y_true, axis=1)

    y_true = tf.dtypes.cast(y_true, tf.int32)
    n_classes = tf.dtypes.cast(5, "float32")
    weights = K.arange(0, n_classes, dtype="float32") / (n_classes - 1)
    weights = (weights - K.expand_dims(weights, -1)) ** 2

    
    act_hist = tf.math.bincount(y_true)
    act_hist = act_hist[1:]
    pred_hist = tf.math.bincount(y_pred)
    pred_hist = pred_hist[1:]
    #act_hist=np.zeros([5])
    #for item in y_true.numpy(): 
    #    act_hist[int(item)]+=1
    
    #pred_hist=np.zeros([5])
    #for item in y_pred.numpy(): 
    #    pred_hist[int(item)]+=1
    
    #tf.convert_to_tensor(act_hist,dtype = "float32")

    #hist_true = tf.convert_to_tensor(act_hist,dtype = "float32")
    #print(hist_true)
    #hist_pred = tf.convert_to_tensor(pred_hist,dtype = "float32")
    #print(hist_pred)
     
    E = K.expand_dims(act_hist, axis=-1) * pred_hist
    #print(E)
    E = E / K.sum(E, keepdims=False)
    #print(E)
    
    O = tf.math.confusion_matrix(y_pred,y_true)
    O = O / K.sum(O)
    #print(O)
    #tf.matmul(weights,O)

    weights= tf.dtypes.cast(weights,"float32")

    O = tf.dtypes.cast(O,"float32")
    E = tf.dtypes.cast(E,"float32")

    num = tf.matmul(weights,O)
    den = tf.matmul(weights,E)

    QWK = (1 - K.sum(num) / K.sum(den))
    return QWK
```

# Metricas

```python
class Metrics(Callback):
    def __init__(self, val_data, batch_size = 32):
        super().__init__()
        self.validation_data = val_data
        self.batch_size = batch_size

    def on_train_begin(self, logs={}):
        self.val_kappas = []

    def on_epoch_end(self, epoch, logs={}):
      
        X_val = self.validation_data[0]
        y_val = self.validation_data[1]
        y_val = y_val.sum(axis=1) - 1
        
        y_pred = self.model.predict(X_val) > 0.5
        y_pred = y_pred.astype(int).sum(axis=1) - 1

        _val_kappa = cohen_kappa_score(
            y_val,
            y_pred, 
            weights='quadratic'
        )

        self.val_kappas.append(_val_kappa)

        print(f"val_kappa: {_val_kappa:.4f}")
        
        if _val_kappa == max(self.val_kappas):
            print("Validation Kappa ha mejorado. Guardando modelo.")
            self.model.save('model.h5')

        return
```

```python
class QWKEvaluation(Callback):
    def __init__(self, validation_data, batch_size=32, interval=1):
        super(Callback, self).__init__()

        self.interval = interval
        self.batch_size = batch_size
        self.y_vals = validation_data
        self.history = []

    def on_epoch_end(self, epoch, logs={}):
        X_val = self.y_vals[0]
        y_val = self.y_vals[1]

        if epoch % self.interval == 0:
            y_pred = self.model.predict(X_val,verbose = 1)

            def flatten(y):
                return np.argmax(y, axis=1).reshape(-1)
            
            score = cohen_kappa_score(flatten(y_val),
                                      flatten(y_pred),
                                      labels=[0,1,2,3,4],
                                      weights='quadratic')
            print("\n epoch: %d - QWK_score: %.6f \n" % (epoch+1, score))
            self.history.append(score)
            if score >= max(self.history):
                print('saving checkpoint: ', score)
                self.model.save('densenet_bestqwk.h5')

qwk = QWKEvaluation(validation_data=(x_val, y_val),
                    batch_size=128, interval=1)
```

```python
model1.load_weights('model.h5')
y_val_pred = model1.predict(x_val)

def compute_score_inv(threshold):
    y1 = y_val_pred > threshold
    y1 = y1.astype(int).sum(axis=1) - 1
    y2 = y_val.sum(axis=1) - 1
    score = cohen_kappa_score(y1, y2, weights='quadratic')
    
    return 1 - score

simplex = scipy.optimize.minimize(
    compute_score_inv, 0.5, method='nelder-mead'
)

best_threshold = simplex['x'][0]

print(best_threshold)
```
