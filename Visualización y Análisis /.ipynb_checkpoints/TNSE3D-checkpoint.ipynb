```python
import numpy as np

aptostraindf = pd.read_csv('/content/drive/My Drive/tesis/aptos/train.csv')
xtrainaptos = np.load('/content/drive/My Drive/tesis/aptos/train224.npy')
```

```python
npix = 224 # resize to npix x npix (for now)
X_train = np.zeros((aptostraindf.shape[0], npix, npix))
```

```python
X_train.shape
```

```python
for i in tqdm_notebook(range(aptostraindf.shape[0])):
    img = xtrainaptos[i]
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    X_train[i, :, :] = cv2.resize(img, (npix, npix)) 
```

```python
print("X_train shape: " + str(np.shape(X_train)))
```

```python
X = X_train / 255

# reshape
X = X.reshape(X.shape[0], -1)
```

```python
trainy=aptostraindf['diagnosis']
```

```python
per_vals = [2,5,10,15,20,30,40,50]

for per in tqdm_notebook(per_vals):
    X_decomposed = TSNE(n_components=3,perplexity = per).fit_transform(X)
    df_tsne = pd.DataFrame(data=X_decomposed, columns=['X','Y','Z'])
    df_tsne['Clase'] = trainy.values
    
    
    fig = px.scatter_3d(df_tsne, x='X', y='Y', z='Z',
              color='Clase', size = 'Clase', opacity=0.7, title = 'Visualizacin Aptos con TSNE')
    fig.show()

    
    #sns.FacetGrid(df_tsne, hue='Score', size=6).map(plt.scatter, 'Dimension_x', 'Dimension_y',).add_legend()
    #plt.title('TSNE for perplexity = ' + str(per))
    #plt.show()
```
